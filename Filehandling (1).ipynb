#11.What is memory management in Python?

   Memory management in Python means how Python handles memory automatically:

  - It allocates memory for variables/objects.

  - Uses reference counting to track object usage.

  - Uses garbage collection to free unused memory.

#Example:

    import gc

    class Person:

    def __init__(self, name):

        self.name = name

    def __del__(self):

        print(f"Object {self.name} destroyed")

# Create object and references
     p1 = Person("Ram")

     p2 = p1  # second reference

     p3 = p1  # third reference

# Delete references one by one
     del p1

     del p2

     print("Two references deleted")

     del p3  # Now all references deleted → __del__ will be called

# Optional: force garbage collection

     gc.collect()


#Output:

   Two references deleted

   Object Ram destroyed

#23.What are the challenges associated with memory management in Python?

 - Memory management in Python involves several challenges, primarily due to its dynamic nature and the way it handles memory allocation and deallocation. Here are some key challenges:

 (i).Garbage Collection: Python uses automatic garbage collection to manage memory. However, circular references (where two or more objects reference each other) can lead to memory leaks if not handled properly.

 (ii).Memory Overhead: Python objects have a certain amount of overhead due to their dynamic nature. This can lead to inefficient memory usage, especially when dealing with a large number of small objects.

 (iii).Fragmentation: Over time, as objects are created and destroyed, memory can become fragmented, leading to inefficient use of memory and potential performance issues.

 (iv).Limited Control: Unlike languages like C or C++, Python does not allow fine-grained control over memory allocation and deallocation, which can be a disadvantage in performance-critical applications.


#Example

     class Node:

     def __init__(self, value):

        self.value = value

        self.next = None

# Create two nodes that reference each other
     node1 = Node(1)

     node2 = Node(2)
     
     node1.next = node2

     node2.next = node1  # Circular reference

# Delete the references
      del node1

      del node2

# At this point, the memory for node1 and node2 is not freed
# because they reference each other, creating a circular reference.

     import gc

# Force garbage collection

     gc.collect()  # This will clean up the circular reference

     print("Garbage collection completed.")


#Output:

     Garbage collection completed.

#25.Why is it important to use multithreading in certain applications?  
  
 - Multithreading is important in certain applications because it allows for concurrent execution of tasks, which can lead to improved performance and responsiveness. Here are some key reasons why multithreading is beneficial:

   (i).Improved Performance: Multithreading can help utilize CPU resources more effectively, especially in I/O-bound applications where threads can perform other tasks while waiting for I/O operations to complete.

   (ii).Responsiveness: In applications with user interfaces, multithreading can keep the UI responsive by offloading long-running tasks to background threads, allowing the main thread to continue processing user input.

   (iii).Parallelism: In CPU-bound applications, multithreading can take advantage of multiple CPU cores to perform computations in parallel, speeding up processing time.

   (iv).Resource Sharing: Threads within the same process share memory and resources, making it easier to share data between them compared to separate processes.

#Example:

   import threading

   import time

  def task1():

    for i in range(5):

        print(f"Task 1 - Count: {i}")

        time.sleep(1)  # Simulate a time-consuming task

     def task2():

     for i in range(5):

        print(f"Task 2 - Count: {i}")

        time.sleep(1)  # Simulate a time-consuming task

# Create threads for each task

     thread1 = threading.Thread(target=task1)

     thread2 = threading.Thread(target=task2)

# Start the threads
     thread1.start()

     thread2.start()

# Wait for both threads to complete

     thread1.join()

     thread2.join()

     print("Both tasks completed.")

#Output
Task 1 - Count: 0

Task 2 - Count: 0

Task 1 - Count: 1

Task 2 - Count: 1

Task 1 - Count: 2

Task 2 - Count: 2

Task 1 - Count: 3

Task 2 - Count: 3

Task 1 - Count: 4

Task 2 - Count: 4

Both tasks completed.
